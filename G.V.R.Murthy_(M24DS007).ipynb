```python
import pandas as pd
for index, row in df.iterrows():
    if df.loc[index, 'quality'] <= 6:
        df.loc[index, 'quality'] = 0
    else:
        df.loc[index, 'quality'] = 1
df.head(10)
```

```python
attributes= df.columns.drop('quality')
for x in attributes:
    min_value = df[x].min()
    max_value = df[x].max()
    df[x] = (df[x] - min_value) / (max_value - min_value)
df.head(10)
```
